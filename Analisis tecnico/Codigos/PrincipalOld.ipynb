```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf
import talib as ta
from csv import reader

#Leer de un archivo CSV el listado de indices a analizar.
with open('C:/Users/alvar/OneDrive/Escritorio/IndicesSYP500.csv', 'r') as archivo:
     leer = reader(archivo)
     lista2D = list(leer)

lista2D.pop(0) #Elimina primer elemento de la lista
        
#Declaramos tickers
tickers = [] 
for item in lista2D:
     #AÃ±adir elementos a la lista 1D
     tickers += item
# Visulaizar resultados de la lista plana o 1D
print(tickers)
```

---------------------------------------------
CREAMOS DATAFRAME CON LA INFO DE PRECIOS

```python
df_precio = pd.DataFrame() #Crear dataframe de precio

for i in tickers:
    df_precio[i] = yf.download(i,start='2020-05-01')['Adj Close']

df_precio.tail()
```

--------------------------------------------------
CALCULAMOS LOS RSI DE TODOS LOS ACTIVOS

```python
df_precio_rsi_200 = pd.DataFrame()
for i in tickers:
    df_precio_rsi_200[i] = ta.RSI(df_precio[i], timeperiod=200)

df_precio_rsi_50 = pd.DataFrame()
for i in tickers:
    df_precio_rsi_50[i] = ta.RSI(df_precio[i], timeperiod=50)

df_precio_rsi_26 = pd.DataFrame()
for i in tickers:
    df_precio_rsi_26[i] = ta.RSI(df_precio[i], timeperiod=26)

df_precio_rsi_14 = pd.DataFrame()
for i in tickers:
    df_precio_rsi_14[i] = ta.RSI(df_precio[i], timeperiod=14)

df_precio_rsi_7 = pd.DataFrame()
for i in tickers:
    df_precio_rsi_7[i] = ta.RSI(df_precio[i], timeperiod=7)   

df_precio_rsi_3 = pd.DataFrame()
for i in tickers:
    df_precio_rsi_3[i] = ta.RSI(df_precio[i], timeperiod=3)

df_precio_rsi_7.tail()
```

AQUI CALCULAMOS LAS MEDIAS DE LOS ACTIVOS (200, 50, 26, 14, 7)

```python
m7 = pd.DataFrame()
m14 = pd.DataFrame()
m26 = pd.DataFrame()
m50 = pd.DataFrame()
m200 = pd.DataFrame()

for i in tickers:
    m7[i]=df_precio[i].rolling(window = 7).mean()
    m14[i]=df_precio[i].rolling(window = 14).mean()
    m26[i]=df_precio[i].rolling(window = 26).mean()
    m50[i]=df_precio[i].rolling(window = 50).mean()
    m200[i]=df_precio[i].rolling(window = 200).mean()
```

---------------------------------------
GRÃFICAS

```python
#AQUÃ DEFiNIMOS LA FUNCIÃN DONDE PODEMOS VISUALIZAR LAS MEDIAS DEL ACTIVO QUE QUERAMOS JUNTO CON SU COTIZACIÃN DIARIA
def plot_precio_medias(prices):

    plt.figure(figsize=(16,6))

    plt.plot(prices.index, prices)

    plt.plot(prices.index,prices.rolling(window = 7).mean())

    plt.plot(prices.index,prices.rolling(window = 14).mean())

    plt.plot(prices.index,prices.rolling(window = 26).mean())

    plt.plot(prices.index,prices.rolling(window = 50).mean())

    plt.plot(prices.index,prices.rolling(window = 200).mean())

    plt.grid()
    plt.title('Cotizaciones diarias y mÃ©dias mÃ³viles', fontsize = 15)
    plt.legend(['Precio','Media 7 precio','Media 14 precio','Media 26 precio','Media 50 precio','Media 200 precio'])
    plt.show()

#Esta parte crea la funciÃ³n para visulaizar la grafica del RSI
def plot_rsi(prices, rsi):
    plt.subplots(figsize=(20,8))
    ax1 = plt.subplot2grid((12,1), (0,0), rowspan = 5, colspan = 1)
    ax2 = plt.subplot2grid((12,1), (5,0), rowspan = 3, colspan = 1)

    ax1.plot(prices)
    ax2.plot(rsi, color = 'grey', linewidth = 1.5, label = "RSI")
    ax2.axhline(y=75, color = "red", linestyle="--")
    ax2.axhline(y=25, color = "green", linestyle="--")

    plt.legend(loc = "lower right")
    plt.show()
```

```python
plot_precio_medias(df_precio['DLTR'])
```

```python
#Con .loc filtro la grÃ¡fica a esa fecha, para que solo aparezca graficada a ese tiempo
#plot_rsi(df_precio['MMM'].loc['2023-05'], df_precio_rsi_7['MMM'].loc['2023-05'])
plot_rsi(df_precio['BIO'], df_precio_rsi_14['BIO'])
```

```python
#En el calculo del MACD debo de ajustar las medias para mi temporalidad que serÃ¡ diaria
data_MACD={}
for i in tickers:
    data_MACD[i] = pd.DataFrame()
    data_MACD[i]['macd'], data_MACD[i]['macd_signal'], data_MACD[i]['macd_hist'] = ta.MACD(df_precio[i], fastperiod=12, slowperiod=26, signalperiod=9)
    data_MACD[i]['adj_close']=df_precio[i]

data_MACD['MMM'].tail
```

```python
#Para graficar el MACDID creamos esta funciÃ³n

def plot_macd(prices, macd, signal, hist):
    plt.subplots(figsize=(20,8))
    ax1 = plt.subplot2grid((12,1), (0,0), rowspan = 5, colspan = 1)
    ax2 = plt.subplot2grid((12,1), (5,0), rowspan = 3, colspan = 1)

    ax1.plot(prices)
    ax2.plot(macd, color = 'grey', linewidth = 1.5, label = "MACD")
    ax2.plot(signal, color = 'skyblue', linewidth = 1.5, label = "SIGNAL")

    for i in range(len(prices)):
        if str(hist[1])[0] == '-':
            ax2.bar(prices.index[i], hist[i], color = '#ef5350')
        else:
            ax2.bar(prices.index[i], hist[i], color = '#26a69a')

    plt.legend(loc = "lower right")
    plt.show()
```

```python
plot_macd(data_MACD['BIO']['adj_close'], data_MACD['BIO']['macd'], data_MACD['BIO']['macd_signal'], data_MACD['BIO']['macd_hist'])
```
